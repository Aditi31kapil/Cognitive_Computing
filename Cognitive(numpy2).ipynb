```python
import numpy as np
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
print('Sum of all elements', np.sum(sum(gfg), axis= 1))
print('Sum of all elements row-wise \n', np.sum(gfg, axis=1))
print('Sum of all elements column-wise', np.sum(gfg, axis=0))
```

```python
#a
array = np.array([10, 52, 62, 16, 16, 54, 453])
print("Sorted array:", np.sort(array))
print("Indices of Sorted array:", np.argsort(array))
print("4 smallest elements:", np.partition(array, 4)[0:4])
print("5 largest elements:", np.partition(array, -5)[-5:])
```

```python
array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
Integer = array[array == np.floor(array)]
print("Integer:" , Integer)
Float = array[array != np.floor(array)]
print("Float:" , Float)
```

```python
# a name= "ADITI KAPIL"
X = ord('A') + ord('K')
print(X)
sales =np.array([X, X+50, X+100, X+150,X+200])
print(sales)
print("--------------------------------------------------")
#b
rate = ((X % 5)+5)/100
taxed_rate = sales * rate
sales_taxed_rate = sales * (rate+1)
print(taxed_rate)
print(sales_taxed_rate)
print("--------------------------------------------------")
#c
discounted_sales = np.where(sales > X + 100, sales * 0.95, sales*0.90)
print(discounted_sales)
print("--------------------------------------------------")
#d
weekly_sales = np.tile(sales, (3,1))
print(weekly_sales)
print("--------------------------------------------------")
weeks = np.arange(3) #to separate the rows
print(weeks)
increased_sales = weekly_sales*(1.02 ** weeks[:,np.newaxis])
print(increased_sales)
print("--------------------------------------------------")
```

```python
import matplotlib.pyplot as plt
def plot_graph(x, y):
    plt.plot(x, y, marker=".", color="red")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.grid()
    plt.show()

x = np.linspace(-10, 10,100)
y1 = x**2
y2 = np.sin(x)
y3 = np.exp(x)
y4 = np.log(np.abs(x)+1)
plt.title("Y = X^2")
plot_graph(x, y1)
plt.title("Y = sin(x)")
plot_graph(x, y2)
plt.title("Y = e^x")
plot_graph(x, y3)
plt.title("Y = log(|x|+1)")
plot_graph(x, y4)
plt.show()
```
