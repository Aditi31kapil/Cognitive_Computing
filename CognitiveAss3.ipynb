```python
import numpy as np
average_salary = np.mean(empdata['Salary'])
bonus = np.sum(empdata['Bonus'])
emp_age = np.min(empdata['Age'])
rate = np.max(empdata['Rating'])
print(average_salary)
print(bonus)
print(emp_age)
print(rate)
```

```python
empdata.sort_values(by = ['Salary'],ascending = False)
```

```python
def categorize_performance(rating):
  if rating >= 4.5:
    return 'Excellent'
  elif rating >= 4.0:
    return 'Good'
  else:
    return 'Average'

empdata['Performance Category'] = empdata['Rating'].apply(categorize_performance)
empdata
```

```python
missing_values = empdata.isnull().sum()
print(missing_values)
```
